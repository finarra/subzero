```python
# import libraries

import pandas as pd
import datetime as datetime
```

```python
# Load files to be merged

weather_to_load = "Resources/weather_frialsa.csv"
alarms_to_load = "Resources/saveris_alarms_complete.csv"

weather_frialsa_df = pd.read_csv(weather_to_load)
alarms_df = pd.read_csv(alarms_to_load)

alarms_df.head()
```

```python

# cpnvert the dt Time Format to Timestamp('2017-03-22 15:16:45')
pd.to_datetime(1622505600, unit='s')

weather_frialsa_df['dt'] = pd.to_datetime(weather_frialsa_df['dt'], unit='s')
weather_frialsa_df.head()

# alarms_df['date'] =  pd.to_datetime(alarms_df['date'], format='%d/%m/%y %H:%M')
# alarms_df.dtypes
```

```python
# Convert the new date object to datetime64
weather_frialsa_df['dt'] = pd.to_datetime(weather_frialsa_df['dt'], format='%d/%m/%y %H:%M')
weather_frialsa_df.dtypes
```

```python
# Rename columns to date so we can merge on 'date' column

weather_frialsa_df = weather_frialsa_df.rename(columns={'dt':'date'})
weather_frialsa_df.head()
```

```python
# Change data type of date in alarms_df to be sure to merge through 'date' with datetime64
alarms_df['date'] =  pd.to_datetime(alarms_df['date'], format='%d/%m/%y %H:%M')
alarms_df.dtypes
```

```python
# Double check shape of pd dfs to be merged

print(f"Alarms df: {alarms_df.shape}")
print(f"Weather df: {weather_frialsa_df.shape}")
```

```python
# Merge both on date time

df_weather_alarms = pd.merge(alarms_df, weather_frialsa_df, on='date', how='left')
print(df_weather_alarms.shape)
df_weather_alarms[6650:6700]
```

```python
# For this case due that we don't have weather data up to may 22 2022 we will need to drop last 450 rows.

# Number of rows to drop
n = 450
 
# Dropping last n rows using drop
df_weather_alarms = df_weather_alarms.drop(df_weather_alarms.tail(n).index,
        inplace = False)
 
# Printing dataframe
df_weather_alarms.shape
```

```python
# We will drop columns we don't need

df_weather_alarms = df_weather_alarms.drop(['temp_min', 'temp_max', 'dt_iso','timezone', 'city_name', 'lat', 'lon', 'visibility', 'dew_point', 'feels_like', 'pressure', 'sea_level', 'grnd_level', 'wind_speed', 'wind_deg', 'wind_gust', 'rain_1h', 'rain_3h', 'snow_3h', 'clouds_all', 'weather_id', 'weather_main', 'weather_description', 'weather_icon', 'snow_1h'], inplace=False, axis=1)

print(df_weather_alarms.shape)

df_weather_alarms.head()
```

```python
# Fill values weather values using fillna()

df_weather_alarms = df_weather_alarms.fillna(method='ffill')

df_weather_alarms = df_weather_alarms.dropna()

print(df_weather_alarms.shape)
df_weather_alarms.head(20)
```

```python
# Check if there are still NaNs
df_weather_alarms.isnull().sum()
```

```python
# Import df as csv
df_weather_alarms.to_csv('Resources/saveris_alarms_weather.csv')
```
