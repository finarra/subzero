```python
# import libraries
import pandas as pd
from datetime import datetime
```

```python
# Visualize Dataframe of Saveris Values
file_to_load = "Resources/saveris_values.csv"
values_df = pd.read_csv(file_to_load, header=0)
print(values_df.shape)
values_df.head()
```

```python
# Combine date and hour column
values_df["Date"] = values_df['date'] + ' ' + values_df['hour']

# Drop column "0" & "1", date and hour
values_df = values_df.drop(['date', 'hour'], inplace=False, axis=1)

# Rename columns
values_df = values_df.rename(columns={'a':'platform_1', 'b':'platform_2', 'c':'deepfreezer_1', 'd':'deepfreezer_2', 'e':'freezer_1', 'f':'freezer_2', 'g':'freezer_3', 'h':'freezer_4', 'i':'cooling_1', 'j':'cooling_2', 'Date':'date'})

# Reorder columns
new_columns_order = ['date','platform_1', 'platform_2', 'deepfreezer_1', 'deepfreezer_2', 'freezer_1', 'freezer_2', 'freezer_3', 'freezer_4', 'cooling_1', 'cooling_2']
values_df = values_df[new_columns_order]

# Drop NaN rows
values_df = values_df.dropna()

# Visualize df
print(values_df.shape)
values_df[31640:]
```

```python
# Check the value types
values_df.dtypes
```

```python
# Convert objects to float
data = ['platform_1', 'deepfreezer_1', 'deepfreezer_2', 'freezer_1', 'freezer_2', 'cooling_1', 'cooling_2']
values_df[data] = values_df[data].astype(float)
values_df.dtypes
```

```python
# Convert the date 'object' type to datetime
values_df['date'] =  pd.to_datetime(values_df['date'], format='%d/%m/%y %H:%M:%S')
values_df.dtypes
```

```python
# Drop index column and visualize the change
values_df.reset_index(drop=True, inplace=True)
values_df.head()
```

```python
# Upload new csv file to Resources folder.
values_df.to_csv('Resources/saveris_values_clean.csv', index=True)
```
