```python
# Import Libraries
import pandas as pd
import datetime as datetime
```

```python
# Alarms dataframe 
file_to_load = "Resources/saveris_alarms.csv"
alarms_df = pd.read_csv(file_to_load)
print(alarms_df.shape)
alarms_df.head()
```

```python
# Drop columns that are not necessary
alarms_df = alarms_df.drop(['Rango','Tipo'], inplace=False, axis=1)

# Rename columns
alarms_df = alarms_df.rename(columns={'Estado':'status', 'Fecha':'date', 'Localidad':'sensor','Grupo de puntos de medición':'location', 'Evento':'alarm_type', 'Value':'value', 'Limit':'limit'})

# New columns order
# Reorder columns
new_columns_order = ['date', 'location', 'sensor', 'status', 'alarm_type', 'limit', 'value']
alarms_df = alarms_df[new_columns_order]

# Drop NaN rows
alarms_df = alarms_df.dropna()

print(alarms_df.shape)
alarms_df.head()
```

```python
# Check data types
alarms_df.dtypes
```

```python
# Change date object type to datetime
alarms_df['date'] =  pd.to_datetime(alarms_df['date'], format='%d/%m/%y %H:%M')
alarms_df.dtypes
```

```python
# create a function
def sensor_type(name):
    if name == 'OCO-Congelacion-Camara2-Sensor1':
        return ('freezer_1')
    elif name == 'OCO-Congelacion-Camara2-Sensor2':
        return ('freezer_2')
    elif name == 'OCO-Anden-Sensor1':
        return ('platform_1')
    elif name == 'OCO-Anden-Sensor2':
        return ('platform_2')
    elif name == 'OCO-Ultracongelador-Camara 3-Sensor1':
        return ('deepfreezer_1')
    elif name == 'OCO-Ultracongelador-Camara 3-Sensor2':
        return ('deepfreezer_2')
    elif name == 'OCO-Refrigeracion-Camara4-Sensor1':
        return ('cooler_1')
    elif name == 'OCO-Refrigeracion-Camara4-Sensor2':
        return ('cooler_2')
# create a new column based on condition
alarms_df['sensor_name'] = alarms_df['sensor'].apply(sensor_type)
# display the dataframe
alarms_df.head()

```
